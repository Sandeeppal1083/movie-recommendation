```python
import ast
import pandas as pd

def convert(obj):
    try:
        if pd.isna(obj):
            return []

        # Only parse if it's a string
        if isinstance(obj, str):
            genre_list = ast.literal_eval(obj)
        elif isinstance(obj, list):  # Already a list
            genre_list = obj
        else:
            return []

        return [item['name'] for item in genre_list if 'name' in item]

    except (TypeError, SyntaxError, ValueError) as e:
        print(f"Error: {e} — Data: {obj}")
        return []

# Apply to column
final_df['genres'] = final_df['genres'].apply(convert)
```

```python
final_df.head(10)
```

```python
final_df['keywords']=final_df['keywords'].apply(convert)
final_df.head(5)
```

```python
import ast
final_df['cast'][0]
def cast_convert(obj):
    L=[]
    counter=0
    if counter!=3:
        genre_list=ast.literal_eval(obj)
        for item in genre_list:
            L.append(item['name'])
            counter+=1
    return L
final_df['cast']=final_df['cast'].apply(cast_convert)
final_df.head(5)
```

```python

def crew_convert(obj):
    L=[]
    crew_list=ast.literal_eval(obj)
    for item in crew_list:
        if item['job']=="Director":
            L.append(item['name'])
    return L
final_df['crew']=final_df['crew'].apply(crew_convert)
final_df['crew'].head(4)
```

```python
final_df['overview']=final_df['overview'].apply(lambda item:item.split())
final_df.head(5)
```

```python
final_df['genres']=final_df['genres'].apply(lambda x: [item.replace(" ","") for item in x])
final_df['cast']=final_df['cast'].apply(lambda x: [item.replace(" ","") for item in x])
final_df['crew']=final_df['crew'].apply(lambda x: [item.replace(" ","") for item in x])
final_df['keywords']=final_df['keywords'].apply(lambda x: [item.replace(" ","") for item in x])
```

```python
final_df['tags']=final_df['overview']+final_df['genres']+final_df['keywords']+final_df['cast']+final_df['crew']
```

```python
final_df.head(2)
```

```python
recom_df=final_df[['id','tags','title']]
recom_df['tags']=recom_df['tags'].apply(lambda x: " ".join(x))
recom_df['tags']=recom_df['tags'].apply(lambda x: x.lower())
```

```python
recom_df.head(3)
```

```python
#from sklearn.feature_extraction.text import CountVectroizer
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
vector=cv.fit_transform(recom_df['tags']).toarray()
cv.get_feature_names_out()
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
def stem_tag(text):
    L=[]
    for item in text.split():
        L.append(ps.stem(item))
    return " ".join(L)
recom_df['tags']=recom_df['tags'].apply(stem_tag)



    
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
sorted(list(enumerate(similarity[2])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    mov_ind=recom_df[recom_df['title']==movie].index[0]
    distance_mov=similarity[mov_ind]
    mov_list=sorted(list(enumerate(distance_mov)),reverse=True,key=lambda x:x[1])[1:6]
    for i in mov_list:
        print(recom_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
